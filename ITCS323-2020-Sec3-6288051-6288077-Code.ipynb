```python
def unreliable_transmission(arr):
    
    noBit = 0 #total bit
    sumBit = 0 
    roundCount = 0  # Count a round.

    size = int(input('size of dataword: '))  # insert size
    p = float(input('p= 0 < p < 1 : '))  # insert p value

    while(roundCount < 100):
        arr = []
        for _ in range(0,size):
            values = randint(0, 2) # random int
            arr.append(values)
        print('\nSender send:',*arr)
        if 0 < p and p < 1:
            sumBit = currupt(p, size, arr)
            noBit = noBit + sumBit
            roundCount = roundCount+1

    print('Total bits:',size*100)
    print('errorneous bits:',noBit)
    print('average ratio of erroneous bits:',round(noBit/(100*size),2))
```

```python
def currupt(p,size,arr):
    noBit = 0 # errorneous bit.
    rp = rand.uniform(0, 1) #random probality
    if rp<=p: # do if rp less than insert prbality value
        random = randint(int(size)/2,int(size)) #random change bit
        for i in range(0,random):
            if(randint(0,2) == 0):
                arr[i] = 1
                noBit += 1
            else:
                arr[i] = 1
                noBit += 1
    print('Reciever : ',*arr,'\n')
    return noBit
```

# 3.CRC

```python
def setCrc(typeCr):
    switcher={
     "CRC-4":"11111",
     "CRC-8":"11101011",
     "Reversed CRC-16":"10100000000000011",
     "CRC-16":"11000000000000101",
     "CRC-24":"1100000000101000100000001",
     "CRC-32":"100000100110000010001110110110111"
    }
    return switcher.get(typeCr,"invalid")
```

```python
def CRC_gen(dataword, word_size, CRC_type):
    if(word_size < 5):
        print("size too low")
    else:
        crcdata = setCrc(CRC_type)
        m = len(crcdata)-1
        newData = dataword
        n = len(dataword)
        for i in range(m):
            newData += "0"

        newData = list(newData)
        for i in range(n):
            if(newData[i] == "1"):
                for j in range(len(crcdata)):
                    if(newData[i+j] == crcdata[j]):
                        newData[i+j] = "0"
                    else:
                        newData[i+j] = "1"
        newData = ''.join(newData)
        newData = newData[n:]
        dataword += newData
        print("codeword:" + dataword)
    return dataword
```

```python
def CRC_check(codeword, CRC_type):
    print("recieved data: " + codeword)
    crcdata = setCrc(CRC_type)
    codeword = list(codeword)
    crc_size = len(crcdata)-1
    n = len(codeword)
    codeword_size = n - crc_size
    for i in range(len(codeword[:codeword_size])):
        if(codeword[i] == "1"):
            for j in range(len(crcdata)):
                if(codeword[i+j] == crcdata[j]):
                    codeword[i+j] = "0"
                else:
                    codeword[i+j] = "1"
    codeword = ''.join(codeword)
    temp = 0
    for i in range(n):
        if(codeword[i] != "0"):
            temp = 1
            break
    if temp == 1: 
        print("FAIL") 
    else:
        print("PASS")
```

```python
def decimalToBinary(n):
    return bin(n).replace("0b", "")      
```

# 5.Hamming Code

```python
def Hamming_gen(codeword):
    j = 0
    k = 1
    m = len(codeword)
    ans = ''
    redun = 0
    for i in range(m):
        if(2**i >= m + i + 1):
            redun = i
            break
    # insert 0 at parity bit
    for i in range(1, m + redun+1):
        if(i == 2**j):
            ans = ans + '0'
            j += 1
        else:
            # add real codeword
            ans = ans + codeword[-1 * k]
            k += 1
  
    # reverse array
    result = ans[::-1]

    n = len(result)

    # find parity bit
    for i in range(redun):
        res = 0
        for j in range(1, n + 1):
  
            # find value of parity bit we found
            if(j & (2**i) == (2**i)):
                res = res ^ int(result[-1 * j])
  
        # concat result
        result = result[:n-(2**i)] + str(res) + result[n-(2**i)+1:]
    return result
```

```python
def Hamming_check(codeword):
    n = len(codeword)
    codeword = list(codeword)
    # reverse data
    codeword = codeword[::-1]
    listParityPos = []
    p = 0

    #find parity bit
    while 2**p <  n+p:
        listParityPos.append(2**p-1)
        p+=1

    p = 0
    wrong = ''
    res= ''
    for i in listParityPos:
        bitOneCount = []
        start = (2**p-1)
        end = (start + 2**p)
        while start < n:    
            # add codeword to end of string
            bitOneCount.extend(codeword[start:end])
            # set start and stop to check value of bit
            start += 2** p + i + 1
            end = start + 2**p
            
        # if odd add 1
        if bitOneCount.count('1') % 2 == 0:
            wrong += '0' 
        else:
            wrong += '1'
        p += 1

    if '1' not in wrong:
        res = 'no error'
    else:
        res = str(int(wrong[::-1],2))
    return res
```
